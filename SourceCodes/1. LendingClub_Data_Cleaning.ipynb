### 2. Load Datasets

#### 2.1. Dataset Loading for LoanStats dataset

Next, we load the LoanStats dataset using the created schema

````python
from datetime import datetime
from pyspark.sql.functions import col, udf
from pyspark.sql.types import DateType

# Load LoanStats dataset
LoanStatsDF_temp = spark.read.csv("./Dataset/LoanStats.csv", header=False, 
                          nullValue="NA", schema=LoanStats_schema)

# User Defined functions for string cell datatype conversion:

# Converts date string of format "MMM-YYYY" to Date datatype
func =  udf (lambda x: datetime.strptime("JAN-1970",'%b-%Y') if x == None else datetime.strptime(x,'%b-%Y'), DateType())
# Converts date string of format "YYYY-MM-DD" to Date datatype
func2 =  udf (lambda x: datetime.strptime("JAN-1970",'%Y-%m-%d') if x == None else datetime.strptime(x,'%Y-%m-%d'), DateType())
# Converts numeric percentage string with "%" suffix to Float datatype
func3 = udf (lambda x: 0.0 if x == None else float(x.strip('%'))/100.0, FloatType())

# Process date string columns in LoanStatsDF dataframe to convert to Date datatype
LoanStatsDF_temp1 = LoanStatsDF_temp.withColumn('issue_d', func(col('issue_d')))
LoanStatsDF_temp2 = LoanStatsDF_temp1.withColumn('earliest_cr_line', func(col('earliest_cr_line')))
LoanStatsDF_temp3 = LoanStatsDF_temp2.withColumn('last_pymnt_d', func(col('last_pymnt_d')))
LoanStatsDF_temp4 = LoanStatsDF_temp3.withColumn('next_pymnt_d', func(col('next_pymnt_d')))
LoanStatsDF_temp5 = LoanStatsDF_temp4.withColumn('last_credit_pull_d', func(col('last_credit_pull_d')))
LoanStatsDF_temp6 = LoanStatsDF_temp5.withColumn('sec_app_earliest_cr_line', func(col('sec_app_earliest_cr_line')))
LoanStatsDF_temp7 = LoanStatsDF_temp6.withColumn('hardship_start_date', func(col('hardship_start_date')))
LoanStatsDF_temp8 = LoanStatsDF_temp7.withColumn('hardship_end_date', func(col('hardship_end_date')))
LoanStatsDF_temp9 = LoanStatsDF_temp8.withColumn('payment_plan_start_date', func(col('payment_plan_start_date')))
LoanStatsDF_temp10 = LoanStatsDF_temp9.withColumn('debt_settlement_flag_date', func(col('debt_settlement_flag_date')))
LoanStatsDF_temp11 = LoanStatsDF_temp10.withColumn('settlement_date', func(col('settlement_date')))

# Process numeric percentage string columns in LoanStatsDF dataframe to convert to Float datatype
LoanStatsDF_temp12 = LoanStatsDF_temp11.withColumn('int_rate', func3(col('int_rate')))
LoanStatsDF = LoanStatsDF_temp12.withColumn('revol_util', func3(col('revol_util')))

# Print out LoanStatsDF schema
LoanStatsDF.printSchema()
````

### 4. Data Cleaning

#### 4.1. Drop columns with >50% missing values

First, we check which columns have more than 50% missing values and drop these columns as they would not have sufficient data for our prediction analysis.

````python
import pyspark.sql.functions as fn
from pyspark.sql.types import StringType
from pyspark.sql.functions import first

# Calculate percentage of missing values for each column in LoanStatsDF dataframe
LoanStatsDF_Missing = LoanStatsDF.agg(*[(1 - fn.count(c) / fn.count("*")).alias(c + '_missing') for c in LoanStatsDF.columns])

# Gather list of columns with >50% missing values in list
columns = [ column for column in LoanStatsDF_Missing.columns if (float(LoanStatsDF_Missing.select(column).first()[0]) >= 0.5) ]

# Print out number of columns with >50% missing values and list out individual missing percentages
print("Number of columns with more than 50% missing values: {0}".format(len(columns)))
for col in columns:
    print(col + " : " + "{:.2%}".format(LoanStatsDF_Missing.select(col).first()[0]) + "\r")

# Print out number of columns in LoanStatsDF dataframe before dropping columns
print("Number of columns in LoanStatsDF before dropping columns with > 50% missing values : " + str(len(LoanStatsDF.columns)))

# Drop columns with >50% missing values
columns_Miss = [s.replace('_missing', '') for s in columns]
print(columns_Miss)
LoanStatsDF_NoMiss = LoanStatsDF.drop(*columns_Miss)

# Print out number of columns in LoanStatsDF dataframe after dropping columns
print("Number of columns in LoanStatsDF after dropping columns with > 50% missing values : " + str(len(LoanStatsDF_NoMiss.columns)))
````

#### 4.2. Drop columns with strong correlations with each other

Next, we check for columns that exhibit strong correlation with each other and drop these columns

````python
import numpy as np

# Extract columns exhibiting high correlation (>0.8)
corr_matrix = LoanStatsDF_NoMiss.toPandas().corr().abs()
high_corr_var=np.where(corr_matrix>0.9)
high_corr_var=[(corr_matrix.columns[x],corr_matrix.columns[y],corr_matrix.iloc[x, y]) for x,y in zip(*high_corr_var) if x!=y and x<y]

high_corr_var.sort(key=lambda tup: tup[2], reverse=True)

s = [[str(e) for e in row] for row in high_corr_var]
lens = [max(map(len, col)) for col in zip(*s)]
fmt = '\t'.join('{{:{}}}'.format(x) for x in lens)
table = [fmt.format(*row) for row in s]
print('\n'.join(table))

# Create empty list for correlated column info
corr_col2 = []

# Extract 2nd correlated column from correlation matrix
for (col1, col2, corrval) in high_corr_var:
    corr_col2.append(col2)

# Get unique values from correlated column info list
corr_col2_uniq = list(set(corr_col2)) 

# Print out number of columns in LoanStatsDF dataframe before dropping highly correlated columns
print("Number of columns in LoanStatsDF before dropping highly correlated columns : " + str(len(LoanStatsDF_NoMiss.columns)))

# Drop highly correlated columns from LoanStatsDF
LoanStatsDF_NoCorr = LoanStatsDF_NoMiss.drop(*corr_col2_uniq)

# Print out number of columns in LoanStatsDF dataframe after dropping highly correlated columns
print("Number of columns in LoanStatsDF after dropping highly correlated columns : " + str(len(LoanStatsDF_NoCorr.columns)))
````

#### 4.6. Check Outlier Data

Next, we check for columns with outliers and drop those columns with high number of outlier data as a percentage of total records and remove the outlier records for those columns with little outlier data

````python
# Extract data type of each column
LoanStatsDF_dtypes = LoanStatsDF_NoNA.dtypes
columns = []

# Extract list of column names of string datatype
for (name, dtype) in LoanStatsDF_dtypes:
    if(dtype != 'string'):
        columns.append(name)

# Empty dict
d = {}
f = {}

# Calculates total number of records
TotalRecords = LoanStatsDF_NoNA.count()

# Extracts the outlier records 1.5 standard deviation from the 10th and 90th percentile
for col in columns:
    quantiles = LoanStatsDF_NoNA.approxQuantile(col, [0.10, 0.90], 0.05)
    IQR = quantiles[1] - quantiles[0]
    bounds = [ quantiles[0] - 1.5 * IQR, quantiles[1] + 1.5 * IQR ]
    LoanStatsDF_outlier = LoanStatsDF_NoNA.filter( (LoanStatsDF_NoNA[col] < bounds[0]) | (LoanStatsDF_NoNA[col] > bounds[1]) )
    d.update({str(col):float(LoanStatsDF_outlier.count() / TotalRecords)})
    f.update({str(col):LoanStatsDF_outlier})
 
# Sorts the outlier record columns by percentage of total records in descending order and display
d_sorted_keys = sorted(d, key=d.get, reverse=True)
for r in d_sorted_keys:
    if(d[r] != 0):
        print("% of outlier records for column " + r + " : {:.2%}".format(d[r]))
        print("Number of outlier records for column " + r + " : {0}".format(f[r].select(r).count()))
        f[r].select(r).orderBy(r, ascending=False).show()
````
